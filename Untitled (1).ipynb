```python
import pandas as pd
import numpy as np
```

```python
# per type => avg total, HP, Attack, Defense
# Ex:          total  attack defense
# Grass      :  20     18      7
# Fire       :  21     19      8

# => read in dataset
# => What types are there
# => Drop unecessary databases
# => Split up into separate databases
# => Calculate mean of variables
# => Save them into a newdataframe
# => Combine all dataframes back together
```

```python
# => read in dataset
pok_df = pd.read_csv("../../code/data ingestion/Pokemon.csv")
```

```python
pok_df
```

```python
# => What types are there
#pok_df
unique_types = pok_df.loc[:,"Type 1"].unique()
unique_types
```

```python
# => Drop unecessary databases
datasets = []
metrics_df = pok_df.loc[:, ["Type 1", "Total", "HP", "Attack", "Defense"]]
for type in unique_types:
    datasets.append(type + "_df")
    
# => Split up into separate databases
file = {}
for type in unique_types:
    file[type] = metrics_df.loc[metrics_df["Type 1"] == type, :]
```

```python
file
```

```python
# => Calculate mean of variables
# => Save them into a newdataframe
# => Combine all dataframes back together
averages = {}
averages_rounded = {}
for x in range(0, len(unique_types)):
    averages_rounded[unique_types[x]] = round(np.mean(file[unique_types[x]].iloc[:,1:]), 0)
    averages[unique_types[x]] = np.mean(file[unique_types[x]].iloc[:,1:])
```

```python
averages
```

```python
# Print the names of the columns.
print ("{:<10} {:<10} {:<10} {:<10} {:<10}".format('Type', 'Avg.Total', 'Avg.HP', 'Avg.Attack', 'Avg.Def'))

# print each data item.
for key, value in averages_rounded.items():
    Type = key
    Total, HP, Attack, Defense = value
    print ("{:<10} {:<10} {:<10} {:<10} {:<10}".format(Type, Total, HP, Attack, Defense))
```

```python
averages_df = pd.DataFrame.from_dict(averages, orient='index')
averages_df
```
